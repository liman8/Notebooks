```python
import requests
from bs4 import BeautifulSoup
import csv

base_url = "https://www.peticije.online/signatures.php?tunnus=podrska_studentima&page_number={}&num_rows=10"

all_names = []

num_pages = 1  # total number of pages as indicated

url = base_url.format(num_pages)
response = requests.get(url)
response.raise_for_status()  # ensure we got a successful response
soup = BeautifulSoup(response.text, 'html.parser')

pagination_div = soup.find('ul', class_='pagination flex-wrap')
if pagination_div:
    # If pagination_div was found, parse within it
    # For example, if the pagination links are inside <a> tags:
    links = pagination_div.find_all('a', href=True)
    num_pages = 1
    from urllib.parse import urlparse, parse_qs
    for link in links:
        href = link['href']
        if "page_number=" in href:
            query = urlparse(href).query
            params = parse_qs(query)
            if 'page_number' in params:
                try:
                    p = int(params['page_number'][0])
                    if p > num_pages:
                        num_pages = p
                except:
                    pass

for page_num in range(1, num_pages + 1):
    url = base_url.format(page_num)
    response = requests.get(url)
    response.raise_for_status()  # ensure we got a successful response
    soup = BeautifulSoup(response.text, 'html.parser')


    # Find the signature table
    signature_table = soup.find('table', id='signatures')
    if not signature_table:
        # If the table isn't found, maybe the structure changed or no data
        continue

    # The first row might be headers; let's skip it
    rows = signature_table.find_all('tr')[1:]

    for row in rows:
        cols = row.find_all('td')
        if len(cols) < 3:
            continue

        full_name = cols[1].get_text(strip=True)
        parts = full_name.split()

        institution = cols[3].get_text(strip=True)

        if len(parts) >= 2:
            first_name = parts[0]
            last_name = " ".join(parts[1:])
        else:
            # If only one part, treat everything as the first name and last name empty
            first_name = full_name
            last_name = ""

        if not institution:
            institution = "N/A"

        all_names.append((first_name, last_name, institution))

# Print out how many names we got
print("Total names scraped:", len(all_names))

# Optional: write to a CSV file
with open("signatories.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["First Name", "Last Name"])
    writer.writerows(all_names)

# If you just want to see a preview
for fn, ln, inst in all_names[:3016]:
    print(fn, ln, " -- ",  inst)
```
